# Clustering

## Definition
**Clustering** is an **unsupervised machine learning** technique used to group similar data points into **clusters** based on their characteristics, similarities, or features.

### Key Points
- No labeled data is required.
- Objects within the same cluster are more similar to each other than to objects in other clusters.
- Used for pattern discovery and data segmentation.

---

# Types of Clustering Algorithms

## 1. Partitioning Methods
These algorithms divide the dataset into a fixed number of clusters.

**Examples:**
- K-Means
- K-Medoids

---

## 2. Hierarchical Methods
These algorithms create a hierarchy (tree) of clusters.

**Types:**
- Agglomerative (Bottom-Up)
- Divisive (Top-Down)

---

## 3. Density-Based Methods
These algorithms form clusters based on the density of data points.

**Example:**
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

---

## 4. Grid-Based Methods
These algorithms divide the data space into a finite number of grid cells and perform clustering on the grid.

**Examples:**
- STING
- CLIQUE

---

## 5. Model-Based Clustering
These algorithms assume that the data is generated from a mixture of underlying probability distributions.

**Examples:**
- Gaussian Mixture Model (GMM)
- Expectation-Maximization (EM)

**Advantages:**
- Can model complex cluster shapes
- Provides probabilistic cluster assignments

---

## 6. Fuzzy Clustering
Unlike hard clustering, each data point can belong to **multiple clusters** with different degrees of membership.

**Example:**
- Fuzzy C-Means (FCM)

**Advantages:**
- Suitable when cluster boundaries overlap
- More flexible than hard clustering

---

# Summary

| Clustering Type | Examples |
|-----------------|----------|
| Partitioning | K-Means, K-Medoids |
| Hierarchical | Agglomerative, Divisive |
| Density-Based | DBSCAN |
| Grid-Based | STING, CLIQUE |
| Model-Based | Gaussian Mixture Model (GMM), EM |
| Fuzzy | Fuzzy C-Means (FCM) |

# Partitioning Clustering (K-Means Clustering)

## Definition
**Partitioning clustering** divides the dataset into a fixed number of clusters (**K**). Each data point belongs to **only one cluster**.

**K-Means Clustering** is one of the most popular partitioning algorithms. It assigns each data point to the cluster whose **centroid** is the nearest.

---

# K-Means Clustering Algorithm

### Step 1: Choose the Number of Clusters (K)
- Assume the number of clusters, denoted by **K**.
- Example: K = 2, K = 3, etc.

---

### Step 2: Initialize the Centroids
- Select **K** initial centroids randomly or based on a chosen method.
- Each centroid represents the center of a cluster.

---

### Step 3: Calculate the Distance
Calculate the distance between each **data point** and every **centroid** using the **Euclidean Distance** formula.

### Euclidean Distance Formula

$$
d = \sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}
$$

For **n-dimensional data**:

$$
d(x,y)=\sqrt{\sum_{i=1}^{n}(x_i-y_i)^2}
$$

where:
- \(x_i\) = Coordinate of the data point
- \(y_i\) = Coordinate of the centroid

---

### Step 4: Assign Data Points
- Compare the distances.
- Assign each data point to the **nearest centroid**.
- Each data point belongs to **only one cluster**.

---

### Step 5: Update the Centroids
After assigning all data points, calculate the **new centroid** of each cluster.

For two points:

$$
C_x=\frac{x_1+x_2}{2}
$$

$$
C_y=\frac{y_1+y_2}{2}
$$
In general, for **n data points**:

$$
C_x=\frac{\sum_{i=1}^{n}x_i}{n}
$$

$$
C_y=\frac{\sum_{i=1}^{n}y_i}{n}
$$

The centroid is simply the **mean (average)** of all the data points in the cluster.

---

### Step 6: Repeat
- Recalculate the distances using the updated centroids.
- Reassign data points if necessary.
- Update the centroids again.
- Continue until:
  - No data point changes its cluster, or
  - The centroids no longer change significantly.

This point is called **convergence**.

---

# Flow of K-Means

```
Choose K
     ↓
Initialize Centroids
     ↓
Calculate Distance
     ↓
Assign Data Points
     ↓
Update Centroids
     ↓
Converged?
   ↓      ↓
 No       Yes
 ↓         ↓
Repeat    Stop
```

---

# Advantages
- Simple and easy to implement.
- Fast for large datasets.
- Efficient when clusters are well separated.

# Disadvantages
- Must specify the value of **K** beforehand.
- Sensitive to outliers.
- Works best with spherical-shaped clusters.
- Different initial centroids may produce different results.

````python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
````

````python
import pandas as pd

data = {
    "Annual_Income": [15,16,17,18,19,20,22,25,27,30,
                      60,62,65,67,70,72,75,77,80,82],
    "Spending_Score": [39,81,6,77,40,76,94,3,72,14,
                       55,52,59,61,58,60,63,65,62,64]
}

df = pd.DataFrame(data)
````

````python
# Create the figure
plt.figure(figsize=(4,3))

# Scatter plot
plt.scatter(df["Annual_Income"], df["Spending_Score"], color="blue")

# Labels and title
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Data")

# Display the plot
plt.show()
````

````python
X = df[["Annual_Income", "Spending_Score"]]
````

````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
scaled_df = pd.DataFrame(
    X_scaled,
    columns=["Annual_Income", "Spending_Score"]
)
````

````python
print(scaled_df.head())
````

````python
# Apply the Elbow Method
wcss = []

for i in range(1, 11):
    kmeans = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(7,5))
plt.plot(range(1,11), wcss, marker='o')

plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")

plt.grid(True)
plt.show()
````

````python
# Train K-Means
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

cluster = kmeans.fit_predict(X_scaled)

# Add cluster labels to the DataFrame
df["Cluster"] = cluster

print(df)
````

# K-Medoids Clustering

## Definition

**K-Medoids** is an **unsupervised machine learning** algorithm used to partition a dataset into **K clusters**.

Unlike **K-Means**, which uses the **mean (centroid)** of the cluster, **K-Medoids** uses an **actual data point** called the **medoid** as the center of the cluster.

A **medoid** is the data point that has the **minimum total distance** to all other points in the same cluster.

---

## Characteristics

- Unsupervised Learning
- Partitioning Clustering Algorithm
- Uses actual data points as cluster centers
- Less sensitive to outliers than K-Means
- Suitable for noisy datasets

---

# Algorithm

### Step 1

Choose the number of clusters.

$$
K = 2
$$

---

### Step 2

Randomly select **K** data points as the initial medoids.

---

### Step 3

Calculate the distance between every data point and each medoid using the Euclidean distance.

$$
d=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}
$$

---

### Step 4

Assign every data point to the nearest medoid.

---

### Step 5

For each cluster,

- Select another data point as a candidate medoid.
- Compute the total distance (cost).
- Replace the current medoid if the total cost decreases.

---

### Step 6

Repeat Steps 3–5 until no medoid changes.

---

# Advantages

- Less sensitive to outliers
- More stable than K-Means
- Uses actual observations as cluster centers

---

# Disadvantages

- Computationally expensive
- Slower than K-Means
- Not suitable for very large datasets
